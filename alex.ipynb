```python
import csv
import pandas as pd
import matplotlib as plt
import ast
import sklearn.linear_model as lm

import numpy as np
import scipy as sp
import scipy.stats as st
import pickle
from sklearn import cross_validation
from sklearn import datasets
from sklearn.metrics import confusion_matrix, roc_curve, auc, f1_score
from sklearn.externals import joblib
from sklearn.naive_bayes import BernoulliNB, MultinomialNB, GaussianNB
from sklearn.cross_validation import train_test_split
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import TfidfVectorizer

import json
import nltk
import unicodedata

from sklearn.feature_extraction.text import CountVectorizer
```

```python
df = pd.DataFrame.from_csv('train.tsv', sep='\t')
for idx, i in enumerate(df.boilerplate):
    try:
        df.boilerplate.iloc[idx] = ast.literal_eval(i)
    except:
        temp = df.boilerplate.iloc[idx].split('null,')
        fixed = ''
        if len(temp) == 2:
            fixed = (temp[0]+'"empty",'+temp[1])
        else:
            fixed = (temp[0]+'"empty",'+temp[1] + '"empty",' + temp[2])
        df.boilerplate.iloc[idx] = ast.literal_eval(fixed)
```

```python
pd.value_counts(df['alchemy_category']).plot(kind='bar');
```

```python
pd.value_counts(df['alchemy_category'][df['label']==1]).plot(kind='bar');
```

```python
ratios = pd.value_counts(df['alchemy_category'][df['label']==1])/pd.value_counts(df['alchemy_category'])
```

```python
labels = np.array(df['label'])
ltest_list = unique(labels)

print ltest_list
```

```python
# vectorizer = CountVectorizer(min_df=1)
vectorizer = TfidfVectorizer(min_df=3,strip_accents=None,ngram_range=(1,2),
                             stop_words='english',max_features=60000,binary=False,
                             norm='l2',use_idf=True,smooth_idf=True,sublinear_tf=True)
# print hydrate['Article Text'][0]
features = []
for i in df.boilerplate:
    features.append(i['body'])
# features.append(df['alchemy_category'])
features = np.array(features)
# compute Bag of words:
labels = np.array(df['label'])
ltest_list = unique(labels)
X = vectorizer.fit_transform(features,y=labels)
# X = vectorizer.fit_transform(features)
x_train, x_split, labels_train, labels_split = train_test_split(X,labels,test_size=0.4)
x_cross, x_test, labels_cross, labels_test = train_test_split(x_split,labels_split,test_size=0.5)


minwrong = 1000
bestalpha = 0
for atest in np.linspace(0,2,101):
    mnb = MultinomialNB(alpha=atest,class_prior=ratios)
    lpred_cross = mnb.fit(x_train,labels_train).predict(x_cross)
    numwrong = (labels_cross != lpred_cross).sum()
    if numwrong < minwrong:
        minwrong = numwrong
        bestalpha = atest
#     print("alpha = %f" % (atest))
#     print("Number of mislabeled points : %d" % numwrong)
print("The best alpha is %f" % (bestalpha))
print("Number of mislabeled points: %d" % minwrong)
```

```python

mnb = MultinomialNB(alpha=bestalpha)
lpred_test = mnb.fit(x_train,labels_train).predict(x_test)
print("alpha = %f" % (bestalpha))
print("Number of mislabeled points : %d" % (lpred_test != labels_test).sum())
f1_mnb = f1_score(labels_test,lpred_test,labels=ltest_list)

cm_mnb = confusion_matrix(labels_test,lpred_test,labels=ltest_list)
fig, ax = plt.subplots(figsize=(10,10))
clb = ax.matshow(cm_mnb,cmap='Blues')
fig.colorbar(clb)
ax.set_xlabel('Predicted label')
ax.set_ylabel('True label')
ax.set_xticks(range(2))
ax.set_yticks(range(2))
ax.set_xticklabels(ltest_list)
ax.set_yticklabels(ltest_list)
ax.set_title('Confusion Matrix');
# plt.tight_layout();
print cm_mnb
```

```python
probs_mnb = mnb.fit(x_train,labels_train).predict_proba(x_test)
fpr_mnb = {}
tpr_mnb = {}
auc_mnb = {}

for i in range(len(ltest_list)):
    print sum(labels_test == ltest_list[i])
    fpr_sk, tpr_sk, thresholds = roc_curve((labels_test==ltest_list[i]), probs_mnb[:,i])
    fpr_mnb[ltest_list[i]] = fpr_sk
    tpr_mnb[ltest_list[i]] = tpr_sk
    try:
        auc_mnb[ltest_list[i]] = auc(fpr_sk,tpr_sk)
    except:
        auc_mnb[ltest_list[i]] = 0
    
    plt.figure(1,figsize=(10,8))
    plt.plot(fpr_mnb[ltest_list[i]],tpr_mnb[ltest_list[i]],label=ltest_list[i])
    plt.legend(loc='best')
for item in auc_mnb.iteritems():
    print item
```
